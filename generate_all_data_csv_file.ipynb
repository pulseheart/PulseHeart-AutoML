## CREATE A CVS FILE WITH THE ENTIRE STUDY DATA-SET

Note: videos labeled TRAIN for Split in the EchoNet source dataset are used for training in our study, those labeled VAL and TEST are used for testing in our experiments. 

```python
import pandas as pd
import numpy as np
import os
```

```python
df_study = pd.read_csv("train_and_test.csv",  index_col= 0)
```

```python
df_study.drop(df_study.columns[[1,2]], axis=1, inplace=True)
df_study=df_study.reset_index()
df_study.rename(columns = {'0':'Videopath', '1':'EFClass'}, inplace = True)
video_list = []
for idx, row in df_study.iterrows():
    name = row.Videopath.split('curated/')[1][:-4]
    video_list.append(name)
df_study['FileName']=video_list
df_study = df_study.iloc[:,[2,1,0]]
df_study = df_study.drop("Videopath", axis=1)
```

```python
df_study.to_csv("df_study.csv", index = False, header = True)       
```
