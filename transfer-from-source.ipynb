```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
df_train = pd.read_csv("annotation datasets/train_L_3_classes.csv", header = None, index_col= 0)
```

```python
df_train
```

```python
df_test = pd.read_csv("annotation datasets/test_L_3_classes.csv", header = None, index_col= 0)
```

```python
df_test
```

```python
df_total = pd.concat([df_train, df_test], ignore_index=False)
```

```python
df_total
```

```python
# check
print(df_total.index[0][68:])
```

## Save results

```python
# transfer all study videos
target = "curated/"
origin = "EchoNet-Dynamic/Videos/"

for ind in df_total.index:
    file_name = ind[68:]
    if os.path.exists(origin + file_name) :
        if not os.path.exists(target + file_name) :
            shutil.copy(origin+file_name, target+file_name)
    else: 
        print(file_name + " not in origin", origin, "not copied")
```
