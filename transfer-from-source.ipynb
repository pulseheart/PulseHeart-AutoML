```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
df_train_and_test = pd.read_csv("train_and_test.csv",  index_col= 0)
```

```python
df_train_and_test
```

```python
# transfer all study videos
target = "curated/"
origin = "EchoNet-Dynamic/Videos/"

for ind in df_train_and_test.index:
    file_name = ind[68:]
    if os.path.exists(origin + file_name) :
        if not os.path.exists(target + file_name) :
            shutil.copy(origin+file_name, target+file_name)
    else: 
        print(file_name + " not in origin", origin, "not copied")
```
